```python
#Import Libraries.
import pandas as pd
import numpy as np
import plotly.io as pio
pio.templates.default = "plotly_dark"
import plotly.graph_objects as go
```

```python
#Import dataset from local host.
data= pd.read_csv('covid_19_data.csv')
data = data.drop(columns='SNo')
data['ObservationDate']= pd.to_datetime(data['ObservationDate'])
data_index = data.set_index('ObservationDate')
data_index.head()
```

```python
#Get a quick overview of the dataset. 
data_index .info()
```

```python
#Computes a summary of statistics pertaining to the dataset columns.
data_index .describe().T
```

```python
# useful aspect of the DatetimeIndex is that the individual date/time components 
#are all available as attributes such as year, month and day.
data_index['Year'] = data_index.index.year
data_index['Month'] = data_index.index.month
data_index['Weekday Name'] = data_index.index.weekday_name
data_index.head()
```

```python
#Indexing with loc, the slice is inclusive of both endpoints.
data_index.loc['2020-03-20':'2020-03-28'].head()
```

```python
data_index.loc["2020-03-27"].sort_values('Confirmed',ascending=False).head()
```

```python
data_index.loc['2020-03'].head()
```

```python
#Group data by Country/Region.
latest_data = data_index.loc['2020-03-27']
country_data = latest_data.groupby(["Country/Region"])[['Confirmed','Deaths','Recovered']].sum().reset_index().sort_values("Confirmed",ascending=0).reset_index(drop=True)
country_data
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=country_data['Country/Region'],
                                y=country_data['Confirmed'],
                                name = 'Confirmed',
                                mode='markers',
                                marker_color=data['Confirmed'],
                                marker = dict(
                                 color = 'rgb(17, 157, 255)',
                                 size = 5,
                                 line = dict(
                                color = 'rgb(231, 99, 250)',
                                width = 2
                               )),
                           ))
                                
fig.update_layout(title='Number of Cases in the World',
                yaxis_type = "log")
fig.show()
```

```python
#Group data by obsevation date.
date_data = data_index.groupby('ObservationDate')[['Confirmed','Deaths','Recovered']].sum().reset_index()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter( x=date_data['ObservationDate'], 
                  y=date_data['Confirmed'],
                  name = 'Confirmed'))

fig.add_trace(go.Scatter( x=date_data['ObservationDate'], 
                  y=date_data['Deaths'],
                  name = 'Deaths'))

fig.add_trace(go.Scatter( x=date_data['ObservationDate'], 
                  y=date_data['Recovered'],
                  name = 'Recovered'))


fig.update_layout(title='Numbers - Confirmed Cases (Worldwide)',
                  xaxis_title="Dates",
                  yaxis_title="Cases",
                  yaxis_type = "log"
                 )
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='white', size=14))

fig.show()
```

## Fluctuation in Observation Numbers (Worldwide)

```python
date_data['Confirmed_Fluct'] = date_data['Confirmed'].diff(periods=1)
date_data['Deaths_Fluct'] = date_data['Deaths'].diff(periods=1)
date_data['Recovered_Fluct'] = date_data['Recovered'].diff(periods=1)
```

```python
date_data.head()
```

```python
date_data = date_data.sort_values(by = 'ObservationDate', ascending = 1)

fig = go.Figure()

fig.add_trace(go.Scatter(x = date_data['ObservationDate'],
                         y = date_data['Confirmed_Fluct'],
                         mode='lines+markers',
                         name = 'Confirmed'))

fig.add_trace(go.Scatter(x = date_data['ObservationDate'],
                         y = date_data['Deaths_Fluct'],
                         mode='lines+markers',
                         name = 'Deaths'))

fig.add_trace(go.Scatter(x = date_data['ObservationDate'],
                         y = date_data['Recovered_Fluct'],
                         mode='lines+markers',
                         name = 'Recovered'))

fig.update_layout(title='Fluctuation - Confirmed, Deaths and Recovered Cases in Worldwide',
                  xaxis = dict(title = 'Observation Dates'),
                  yaxis = dict(title = 'Number of Cases'),
                  yaxis_type = "log")

fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='white', size=14))

fig.show()
```

## Fluctuation in Observation Numbers (Malaysia)

```python
Malaysia = data_index[data_index["Country/Region"] == 'Malaysia'].reset_index()
Mly_data = Malaysia.groupby('ObservationDate')[['Confirmed','Deaths','Recovered']].sum().reset_index()

Mly_data['Confirmed_Fluct'] = Mly_data['Confirmed'].diff(periods=1)
Mly_data['Deaths_Fluct'] = Mly_data['Deaths'].diff(periods=1)
Mly_data['Recovered_Fluct'] = Mly_data['Recovered'].diff(periods=1)

date_data = date_data.sort_values(by = 'ObservationDate', ascending = 1)

fig = go.Figure()
fig.add_trace(go.Scatter(x = Mly_data['ObservationDate'],
                         y = Mly_data['Confirmed_Fluct'],
                         mode='lines+markers',
                         name = 'Confirmed'))

fig.add_trace(go.Scatter(x = Mly_data['ObservationDate'],
                         y = Mly_data['Deaths_Fluct'],
                         mode='lines+markers',
                         name = 'Deaths'))

fig.add_trace(go.Scatter(x = Mly_data['ObservationDate'],
                         y = Mly_data['Recovered_Fluct'],
                         mode='lines+markers',
                         name = 'Recovered'))

fig.update_layout(title='Fluctuation - Confirmed, Deaths and Recovered Cases in Malaysia',
                  xaxis = dict(title = 'ObservationDate'),
                  yaxis = dict(title = 'Number of Cases'),
                  yaxis_type = "log")
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='white', size=14))
fig.show()
```
